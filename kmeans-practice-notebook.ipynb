# Practicing Kmeans Clustering using sklearn - make_blobs

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std= 1, random_state = 35)
```

```python
plt.scatter(df[0][:,0], df[0][:,1], c = df[1])
```

```python
data = pd.DataFrame(df[0], columns = ['1','2'])
data['labels'] = df[1]
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[['1', '2']])
```

```python
kmeans.labels_
```

```python
predictions = kmeans.labels_
```

```python
kmeans.inertia_ # This is same as SSE
```

```python
kmeans.cluster_centers_
```

```python
SSE = []

for i in range(1,10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data[['1', '2']])
    SSE.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), SSE, marker = 'o')
```

```python
f, a = plt.subplots(1,2, sharey = True, figsize = (10,5))
a[0].scatter(df[0][:,0], df[0][:,1], c = df[1])
a[0].set_title('Original')
a[1].scatter(df[0][:,0], df[0][:,1], c = predictions)
a[1].set_title('Predicted')
```
